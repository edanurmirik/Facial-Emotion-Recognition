```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model

model_path = 'model.keras'
model = load_model(model_path)

emotion_labels = ['Kizgin', 'Igrenmis', 'Korku', 'Mutlu', 'Uzgun', 'Saskin', 'Notr']

emotion_colors = {
    'Kizgin': (0, 0, 255),       # Kırmızı
    'Igrenmis': (0, 165, 255),   # Turuncu
    'Korku': (128, 0, 128),      # Mor
    'Mutlu': (0, 255, 0),        # Yeşil
    'Uzgun': (255, 0, 0),        # Mavi
    'Saskin': (0, 255, 255),     # Sarı
    'Notr': (255, 255, 255)      # Beyaz
}

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)

    for (x, y, w, h) in faces:
        roi_gray = gray[y:y+h, x:x+w]
        roi_gray_resized = cv2.resize(roi_gray, (48, 48))
        roi_gray_resized = roi_gray_resized / 255.0
        roi_gray_resized = np.expand_dims(roi_gray_resized, axis=0)  
        roi_gray_resized = np.expand_dims(roi_gray_resized, axis=-1) 
        #roi_gray_resized = np.repeat(roi_gray_resized, 3, axis=-1)  

        prediction = model.predict(roi_gray_resized)
        emotion_index = np.argmax(prediction)
        emotion = emotion_labels[emotion_index]

        color = emotion_colors.get(emotion, (255, 255, 255)) 
        cv2.rectangle(frame, (x, y), (x+w, y+h), color, 2)
        cv2.putText(frame, emotion, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, color, 2)

    cv2.imshow('Emotion Detection', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
